<span style="generic-family:Arial; font-size:1.0em;color:black;"> 1. Print out a list of contenst of the wikipedia website about Dachshund 

<img src="https://upload.wikimedia.org/wikipedia/commons/6/61/Dackel_mit_Bierkrug.jpg" width='200' height='200'>


```python
from bs4 import BeautifulSoup
import requests
url='https://en.wikipedia.org/wiki/Dachshund'
# use requests to load html from the webpage into Python
r=requests.get(url)
#
html_doc=r.text
# use BeautifulSoup to parser the html
soup=BeautifulSoup(html_doc,features="html.parser") # add features="html.parser"
# inspect element to find the html tags that contain all the titles 
# extract all the titles from the html using BeautifulSoup
title = soup.find_all(class_='mw-headline')
for headline in title:
    print(headline.text)
```

<span style="generic-family:Arial; font-size:1.0em;color:black;">2. Find all the links from the New York Times homepage

```python
from bs4 import BeautifulSoup
import requests
url='https://en.wikipedia.org/wiki/Dachshund'
# use requests to load html from the webpage into Python
r=requests.get(url)
#
html_doc=r.text
# use BeautifulSoup to parser the html
soup=BeautifulSoup(html_doc,features="html.parser") # add features="html.parser"
# inspect element to find the html tags that contain all the titles 
# extract all the links from the html using BeautifulSoup
# filter href as is it an attribute that specifies the destination address of the link 
#also ingnore the lines that have only '/' 
link = soup.find_all('a', href=True)
for links in link:
    if links['href'].startswith('http'):
        print(links['href'])
```

<span style="generic-family:Arial; font-size:1.0em;color:black;"> 3. Print out the text from the wepage 

```python
from bs4 import BeautifulSoup
import requests
url='https://en.wikipedia.org/wiki/Dachshund'
r=requests.get(url)
html_doc=r.text
soup=BeautifulSoup(html_doc,features="html.parser") # add features="html.parser"
# by inspecting the elements we can observe that the p appears when there is a text, therefore let's find all 'p'
texts= soup.find_all('p')

for text in texts:
    content=str(text.getText())
    print(content)
```
